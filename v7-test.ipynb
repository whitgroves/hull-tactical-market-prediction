```python
import os
import pandas as pd
import numpy as np
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import keras
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('test.csv').fillna(0).drop(columns=['date_id', 'is_scored', 'lagged_risk_free_rate', 'lagged_forward_returns'])
data
```

```python
y = np.where(data['lagged_market_forward_excess_returns'] > 0, 2.0, 0.0)
y
```

```python
X = data.drop(columns=['lagged_market_forward_excess_returns'])
columns = X.columns
for col in columns:
    X[f'lagged_{col}'] = X[col].shift(1)
X = X.fillna(0)
X
```

```python
models = []
for file in os.listdir('.models'):
    if file.startswith('v7'):
        model = keras.saving.load_model(f'.models/{file}')
        models.append(model)
```

```python
for model in models:
    plt.figure(figsize=(13, 5))
    plt.title(f'{model.name} vs Hull data')
    plt.plot(X.index, y[1:])
    plt.scatter(X.index, y[1:], label='Targets', edgecolors='k', c='g', s=64)
    predictions = model.predict(X)
    plt.scatter(X.index, predictions, label='Predictions', marker='X', edgecolors='k', c='orange', s=64)
    plt.legend()
```

```python
for model in models:
    print(model.predict(X)[-1][0].item())
```
